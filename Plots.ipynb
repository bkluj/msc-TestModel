## Training loss

```python
plt.figure(figsize=(10, 6))
plt.plot(df_MB2["epoch"], df_MB2["train_loss"], label='MobileNetV2', linestyle='-', marker='o')
plt.plot(df_RN18["epoch"], df_RN18["train_loss"], label='ResNet18', linestyle='--', marker='x')
plt.plot(df_VGG16["epoch"], df_VGG16["train_loss"], label='VGG16', linestyle='-.', marker='s')

# Corrected labels using train_loss
last_ep_MB2 = df_MB2["epoch"].iloc[-1]
last_val_MB2 = df_MB2["train_loss"].iloc[-1]
plt.text(last_ep_MB2 + 0.5, last_val_MB2, f'{last_val_MB2:.3f}', color='tab:blue', fontsize=9)

last_ep_RN18 = df_RN18["epoch"].iloc[-1]
last_val_RN18 = df_RN18["train_loss"].iloc[-1]
plt.text(last_ep_RN18 + 0.5, last_val_RN18, f'{last_val_RN18:.3f}', color='tab:orange', fontsize=9)

last_ep_VGG16 = df_VGG16["epoch"].iloc[-1]
last_val_VGG16 = df_VGG16["train_loss"].iloc[-1]
plt.text(last_ep_VGG16 + 0.5, last_val_VGG16, f'{last_val_VGG16:.3f}', color='tab:green', fontsize=9)

plt.xlabel('Epoka')
plt.ylabel('Strata treningowa')
plt.title('Porównanie strat treningowych w danej epoce dla różnych architektur')
plt.legend()
plt.grid(True)
plt.tight_layout()


plt.savefig("Resources/model_comparison_train_loss.svg", format="svg")
plt.show()
```

## Validation loss

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSVs
df_MB2 = pd.read_csv("MobilenetV2-Resources/msc-MobilenetV2History.csv")
df_RN18 = pd.read_csv("ResNet18-Resources/msc-ResNet18History.csv")
df_VGG16 = pd.read_csv("VGG16-Resources/msc-VGG16History.csv")

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df_MB2["epoch"], df_MB2["val_loss"], label='MobileNetV2', linestyle='-', marker='o')
plt.plot(df_RN18["epoch"], df_RN18["val_loss"], label='ResNet18', linestyle='--', marker='x')
plt.plot(df_VGG16["epoch"], df_VGG16["val_loss"], label='VGG16', linestyle='-.', marker='s')

# Add last-value labels
last_ep_MB2 = df_MB2["epoch"].iloc[-1]
last_val_MB2 = df_MB2["val_loss"].iloc[-1]
plt.text(last_ep_MB2 + 0.5, last_val_MB2, f'{last_val_MB2:.3f}', color='tab:blue', fontsize=9)

last_ep_RN18 = df_RN18["epoch"].iloc[-1]
last_val_RN18 = df_RN18["val_loss"].iloc[-1]
plt.text(last_ep_RN18 + 0.5, last_val_RN18, f'{last_val_RN18:.3f}', color='tab:orange', fontsize=9)

last_ep_VGG16 = df_VGG16["epoch"].iloc[-1]
last_val_VGG16 = df_VGG16["val_loss"].iloc[-1]
plt.text(last_ep_VGG16 + 0.5, last_val_VGG16, f'{last_val_VGG16:.3f}', color='tab:green', fontsize=9)

# Labels and legend
plt.xlabel('Epoka')
plt.ylabel('Strata walidacyjna')
plt.title('Porównanie strat walidacyjnych w danej epoce dla różnych architektur')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.savefig("Resources/model_comparison_val_loss.png")
```

## Training accuracy

```python
plt.figure(figsize=(10, 6))
plt.plot(df_MB2["epoch"], df_MB2["train_accuracy"], label='MobileNetV2', linestyle='-', marker='o')
plt.plot(df_RN18["epoch"], df_RN18["train_accuracy"], label='ResNet18', linestyle='--', marker='x')
plt.plot(df_VGG16["epoch"], df_VGG16["train_accuracy"], label='VGG16', linestyle='-.', marker='s')

last_ep_MB2 = df_MB2["epoch"].iloc[-1]
last_val_MB2 = df_MB2["train_accuracy"].iloc[-1]
plt.text(last_ep_MB2 + 0.5, last_val_MB2, f'{last_val_MB2:.3f}', color='tab:blue', fontsize=9)

last_ep_RN18 = df_RN18["epoch"].iloc[-1]
last_val_RN18 = df_RN18["train_accuracy"].iloc[-1]
plt.text(last_ep_RN18 + 0.5, last_val_RN18, f'{last_val_RN18:.3f}', color='tab:orange', fontsize=9)

last_ep_VGG16 = df_VGG16["epoch"].iloc[-1]
last_val_VGG16 = df_VGG16["train_accuracy"].iloc[-1]
plt.text(last_ep_VGG16 + 0.5, last_val_VGG16, f'{last_val_VGG16:.3f}', color='tab:green', fontsize=9)

plt.xlabel('Epoka')
plt.ylabel('Dokładność treningowa')
plt.title('Porównanie dokładności treningowych w danej epoce dla różnych architektur')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.savefig("Resources/model_comparison_train_acc.png")
```

## Validation accuracy

```python
plt.figure(figsize=(10, 6))
plt.plot(df_MB2["epoch"], df_MB2["val_accuracy"], label='MobileNetV2', linestyle='-', marker='o')
plt.plot(df_RN18["epoch"], df_RN18["val_accuracy"], label='ResNet18', linestyle='--', marker='x')
plt.plot(df_VGG16["epoch"], df_VGG16["val_accuracy"], label='VGG16', linestyle='-.', marker='s')

last_ep_MB2 = df_MB2["epoch"].iloc[-1]
last_val_MB2 = df_MB2["val_accuracy"].iloc[-1]
plt.text(last_ep_MB2 + 0.5, last_val_MB2, f'{last_val_MB2:.3f}', color='tab:blue', fontsize=9)

last_ep_RN18 = df_RN18["epoch"].iloc[-1]
last_val_RN18 = df_RN18["val_accuracy"].iloc[-1]
plt.text(last_ep_RN18 + 0.5, last_val_RN18, f'{last_val_RN18:.3f}', color='tab:orange', fontsize=9)

last_ep_VGG16 = df_VGG16["epoch"].iloc[-1]
last_val_VGG16 = df_VGG16["val_accuracy"].iloc[-1]
plt.text(last_ep_VGG16 + 0.5, last_val_VGG16, f'{last_val_VGG16:.3f}', color='tab:green', fontsize=9)

plt.xlabel('Epoka')
plt.ylabel('Dokładność walidacyjna')
plt.title('Porównanie dokładności walidacyjnych w danej epoce dla różnych architektur')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.savefig("Resources/model_comparison_val_acc.svg")
```

```python
# Function to extract final metrics from a DataFrame
def get_final_metrics(df, model_name):
    last = df.iloc[-1]
    return {
        "Model": model_name,
        "Train Loss": last["train_loss"],
        "Train Accuracy": last["train_accuracy"],
        "Validation Loss": last["val_loss"],
        "Validation Accuracy": last["val_accuracy"]
    }

# Build summary list
summary_data = [
    get_final_metrics(df_MB2, "MobileNetV2"),
    get_final_metrics(df_RN18, "ResNet18"),
    get_final_metrics(df_VGG16, "VGG16")
]

# Create comparison DataFrame
comparison_df = pd.DataFrame(summary_data)
comparison_df = comparison_df.round(4)

# Display the result
print(comparison_df)
```

```python
def add_last_value_label(x, y, label_color='black'):
    plt.text(x[-1] + 0.5, y[-1], f'{y[-1]:.3f}', fontsize=9, color=label_color)
```
